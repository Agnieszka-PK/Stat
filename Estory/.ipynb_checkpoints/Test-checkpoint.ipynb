```python
import matplotlib.pyplot as plt
plt.rc('pgf', texsystem='pdflatex')
x = range(0, 10)
y = [t**2 for t in x]
z = [t**2+1 for t in x]
plt.plot(x, y, dpi=400)
plt.plot(x, z, label=r'$\beta=\alpha^2+1$')
plt.xlabel(r'$\alpha$')
plt.ylabel(r'$\beta$')
plt.legend(loc=0)
plt.savefig('C:\\Users\\123\\Documents\\DDR\\Praca\\t1.pgf')
```
